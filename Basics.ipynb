# Exploring Hacker News Posts

In this project, we’ll analyze a dataset of submissions from [Hacker News](https://news.ycombinator.com), a popular technology and startup discussion site. The [dataset](https://www.kaggle.com/datasets/hacker-news/hacker-news-posts/data) contains around 20,000 posts, each with details such as title, author, points, number of comments, and submission time.
Our main goal is to compare two types of posts — Ask HN (where users ask questions) and Show HN (where users share projects or ideas) — to determine which type tends to receive more comments on average and whether the time a post is created affects the level of engagement.

```python
from csv import reader
opened_file = open('hacker_news.csv')
read_file = reader(opened_file)
hn = list(read_file)
opened_file.close()
print(hn[:5])
```

```python
headers = hn[0]
hn = hn[1:]
```

```python
ask_posts, show_posts, other_posts = list(), list(), list()

for post in hn:
    title = post[1]
    if title.lower().startswith("ask hn"):
        ask_posts.append(post)
    elif title.lower().startswith("show hn"):
        show_posts.append(post)
    else:
        other_posts.append(post)

print("The number of ask posts is {0}".format(len(ask_posts)))
print("The number of show posts is {0}".format(len(show_posts)))
print("The number of other posts is {0}".format(len(other_posts)))
```

```python
total_ask_comments = 0
for post in ask_posts:
    comments = int(post[4])
    total_ask_comments += comments
avg_ask_comments = total_ask_comments/len(ask_posts)
print("Average ask comment: ", avg_ask_comments)

total_show_comments = 0
for post in show_posts:
    comments = int(post[4])
    total_show_comments += comments
avg_show_comments = total_show_comments/len(show_posts)
print("Average show comment: ",avg_show_comments)
```

We're able to determine that ask posts receive more comments on average

Next, we'll determine if ask posts created at a certain time are more likely to attract comments. We'll use the following steps to perform this analysis:
 - 

```python
from datetime import datetime as dt

result_list = list()

for post in ask_posts:
    created_at = post[6]
    comments = int(post[4])
    create_comment = [created_at, comments]
    result_list.append(create_comment)

counts_by_hour, comments_by_hour = dict(), dict()

for row in result_list:
    datetime = dt.strptime(row[0],"%m/%d/%Y %H:%M")
    hour = datetime.strftime("%H")
    if hour in counts_by_hour:
        counts_by_hour[hour] += 1
        comments_by_hour[hour] += row[1] 
    else:
        counts_by_hour[hour] = 1
        comments_by_hour[hour] = row[1] 
print("The number of comments by hour: ", comments_by_hour, '\n')
print("The number of counts by hour: ", counts_by_hour, '\n')
```

```python
avg_by_hour = list()
swap_avg_by_hour = list()

for hour in comments_by_hour:
    average = comments_by_hour[hour]/ counts_by_hour[hour]
    avg_by_hour.append([hour, average])
    swap_avg_by_hour.append([average, hour])
print(avg_by_hour)
```

```python
sorted_swap = sorted(swap_avg_by_hour, reverse=True)

print("Top 5 Hours for Ask Posts Comments: ")
for average in sorted_swap[:5]:
    time = dt.strptime(average[1],'%H')
    time = time.strftime("%H:%M")
    
    print("{0}: {1:.2f} average comments per post".format(time, average[0]))
```
